```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU, Flatten, Activation
from tensorflow.keras.initializers import RandomNormal, RandomUniform, HeNormal, HeUniform, Orthogonal
from tensorflow.keras.optimizers import Adam, Adagrad, RMSprop, SGD
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.datasets import mnist

# DATA LOADING & PREPROCESSING 

(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize to [0,1] for better training stability
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

#  MODEL FACTORY FUNCTION

def build_model(initializer, activation_type):
    """
    initializer   : Keras initializer instance (RandomNormal, HeNormal, etc.)
    activation_type: 'tanh' or 'leaky_relu'
    """
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))

    # Hidden layer configs: (units, dropout_rate, use_batchnorm)
    hidden_layers = [
        (128, 0.3, True),
        (64, 0.3, True),
        (32, 0.2, True),
        (16, 0.1, False),
    ]

    for units, dropout_rate, use_bn in hidden_layers:
        model.add(Dense(units, kernel_initializer=initializer))

        if activation_type == "leaky_relu":
            model.add(LeakyReLU(negative_slope=0.1))
        else:
            model.add(Activation(activation_type))

        if use_bn:
            model.add(BatchNormalization())

        model.add(Dropout(dropout_rate))

    # Output layer
    model.add(Dense(10, activation='softmax'))

    return model

# ===================== EXPERIMENT CONFIGS ================================

configs = [
    {
        "name": "RandomNormal + LeakyReLU + Adam",
        "initializer": RandomNormal(mean=0.0, stddev=1.0),
        "activation": "leaky_relu",
        "optimizer": Adam(),
        "patience": 10,
    },
    {
        "name": "RandomUniform + Tanh + Adam",
        "initializer": RandomUniform(minval=0.0, maxval=1.0),
        "activation": "tanh",
        "optimizer": Adam(),
        "patience": 10,
    },
    {
        "name": "HeNormal + Tanh + Adagrad(0.01)",
        "initializer": HeNormal(),
        "activation": "tanh",
        "optimizer": Adagrad(learning_rate=0.01),
        "patience": 10,
    },
    {
        "name": "HeUniform + Tanh + RMSprop(0.001)",
        "initializer": HeUniform(),
        "activation": "tanh",
        "optimizer": RMSprop(learning_rate=0.001),
        "patience": 10,
    },
    {
        "name": "Orthogonal + Tanh + SGD(0.01)",
        "initializer": Orthogonal(),
        "activation": "tanh",
        "optimizer": SGD(learning_rate=0.01),
        "patience": 6,
    },
]

histories = {}
test_results = {}

# ===================== TRAINING LOOP FOR ALL CONFIGS ====================

for cfg in configs:
    name = cfg["name"]
    print(f"\n\n================= Training: {name} =================")

    model = build_model(cfg["initializer"], cfg["activation"])
    model.compile(
        optimizer=cfg["optimizer"],
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    model.summary()

    early_stop = EarlyStopping(
        monitor="val_loss",
        patience=cfg["patience"],
        restore_best_weights=True
    )

    history = model.fit(
        x_train, y_train,
        validation_split=0.2,
        epochs=16,
        batch_size=32,
        verbose=1,
        callbacks=[early_stop]
    )

    histories[name] = history

    # Evaluate on test data
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    test_results[name] = (test_loss, test_acc)
    print(f"[{name}] Test Loss: {test_loss:.4f} | Test Accuracy: {test_acc:.4f}")

# ===================== PLOTTING COMPARISONS =============================

# 1. Validation accuracy comparison
plt.figure(figsize=(8, 6))
for name, history in histories.items():
    plt.plot(history.history["val_accuracy"], label=name)
plt.title("Validation Accuracy per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Validation loss comparison
plt.figure(figsize=(8, 6))
for name, history in histories.items():
    plt.plot(history.history["val_loss"], label=name)
plt.title("Validation Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ===================== PRINT FINAL TEST METRICS TABLE ===================

print("\n\n====== Final Test Metrics ======")
for name, (loss, acc) in test_results.items():
    print(f"{name:40s}  |  Loss: {loss:.4f}  |  Accuracy: {acc:.4f}")
```
