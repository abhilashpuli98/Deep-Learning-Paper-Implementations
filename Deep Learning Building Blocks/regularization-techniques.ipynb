```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU, Flatten, Activation
from tensorflow.keras.initializers import RandomNormal, RandomUniform, HeNormal, HeUniform, Orthogonal
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam, Adagrad, RMSprop, SGD
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.datasets import mnist

# DATA LOADING & PREPROCESSING
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0


# MODEL FACTORY FUNCTION — NOW WITH L2 REGULARIZATION
def build_model(initializer, activation_type):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))

    hidden_layers = [
        (128, 0.3, True),
        (64, 0.3, True),
        (32, 0.2, True),
        (16, 0.1, False),
    ]

    for units, dropout_rate, use_bn in hidden_layers:
        model.add(Dense(
            units,
            kernel_initializer=initializer,
            kernel_regularizer=l2(0.001)  # <---- L2 Regularization added here
        ))

        if activation_type == "leaky_relu":
            model.add(LeakyReLU(negative_slope=0.1))
        else:
            model.add(Activation(activation_type))

        if use_bn:
            model.add(BatchNormalization())

        model.add(Dropout(dropout_rate))

    # Output layer
    model.add(Dense(10, activation="softmax"))

    return model


# EXPERIMENT CONFIGS
configs = [
    {"name": "RandomNormal + LeakyReLU + Adam", "initializer": RandomNormal(mean=0.0, stddev=1.0), "activation": "leaky_relu", "optimizer": Adam(), "patience": 10},
    {"name": "RandomUniform + Tanh + Adam", "initializer": RandomUniform(0.0, 1.0), "activation": "tanh", "optimizer": Adam(), "patience": 10},
    {"name": "HeNormal + Tanh + Adagrad(0.01)", "initializer": HeNormal(), "activation": "tanh", "optimizer": Adagrad(0.01), "patience": 10},
    {"name": "HeUniform + Tanh + RMSprop(0.001)", "initializer": HeUniform(), "activation": "tanh", "optimizer": RMSprop(0.001), "patience": 10},
    {"name": "Orthogonal + Tanh + SGD(0.01)", "initializer": Orthogonal(), "activation": "tanh", "optimizer": SGD(0.01), "patience": 6},
]

histories = {}
test_results = {}

# TRAINING LOOP
for cfg in configs:
    print(f"\n\n================= Training: {cfg['name']} =================")

    model = build_model(cfg["initializer"], cfg["activation"])
    model.compile(loss="sparse_categorical_crossentropy", optimizer=cfg["optimizer"], metrics=["accuracy"])

    model.summary()

    early_stop = EarlyStopping(monitor="val_loss", patience=cfg["patience"], restore_best_weights=True)

    history = model.fit(
        x_train, y_train,
        validation_split=0.2,
        epochs=16,
        batch_size=32,
        verbose=1,
        callbacks=[early_stop]
    )

    histories[cfg["name"]] = history

    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    test_results[cfg["name"]] = (test_loss, test_acc)
    print(f"[{cfg['name']}] Test Loss: {test_loss:.4f} | Test Accuracy: {test_acc:.4f}")


# PLOT: Validation Accuracy
plt.figure(figsize=(8, 6))
for name, hist in histories.items():
    plt.plot(hist.history["val_accuracy"], label=name)
plt.title("Validation Accuracy per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# PLOT: Validation Loss
plt.figure(figsize=(8, 6))
for name, hist in histories.items():
    plt.plot(hist.history["val_loss"], label=name)
plt.title("Validation Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
plt.grid(True)
plt.show()

# FINAL TEST RESULTS TABLE
print("\n====== Final Test Metrics ======")
for name, (loss, acc) in test_results.items():
    print(f"{name:40s} | Loss: {loss:.4f} | Accuracy: {acc:.4f}")
```
